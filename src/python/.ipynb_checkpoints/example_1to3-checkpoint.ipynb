```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('../../data/tips_data.csv')
```

```python
data.head()
```

## Example 1

```python
sns.set(style='white', font_scale=1.7)
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(x=data['total_bill'], y=data['tip'])
ax.set_xlabel('Total Bill (£)')
ax.set_ylabel('Tip (£)')
sns.despine()
plt.show()
```

## Example 2

```python
sns.set(style='white', font_scale=1.7)
fig, ax = plt.subplots(figsize=(10,10))
sns.regplot(x='total_bill', y='tip', data=data)
ax.set_xlabel('Total Bill (£)')
ax.set_ylabel('Tip (£)')
sns.despine()
plt.show()
```

## Example 3

```python
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True) # initiate plotly notebook mode
```

```python
male = go.Scatter( x = data[data.sex=='Male']['total_bill'], y = data[data.sex=='Male']['tip'], name='Male', text=data['smoker'], mode='markers', marker = {'size':12, 'color':'red'})
female = go.Scatter( x = data[data.sex=='Female']['total_bill'], y = data[data.sex=='Female']['tip'], name='Female', text=data['smoker'], mode='markers', marker = {'size':12, 'color':'blue'})

iplot([male,female])
```
